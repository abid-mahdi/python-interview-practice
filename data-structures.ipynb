# 1. Lists

```python
animals = ['cat', 'dog', 'zebra', 'donkey']

# index - finds first occurence with optional range. O(n) time
try:
  animals.index('cats', 0, 2)
except ValueError as ex:
  print('Value does not exist in the list: ' + str(ex))
```

```python
# append - O(1) time
animals.append('frog')
animals
```

```python
# extend - use to add all the lists elements to a list. O(k) time where k is elements to be added
animals_tuple = ('hamster', 'pig')
animals_set = {'tiger'}
animals_list = ['kangaroo', 'griaffe']

animals.extend(animals_tuple)
animals.extend(animals_set)
animals.extend(animals_list)
animals
```
